```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import joblib

df=pd.read_csv('wind_dataset.csv')


numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())


df["DATE"] = pd.to_datetime(df["DATE"])
df["Year"] = df["DATE"].dt.year
df["Month"] = df["DATE"].dt.month
df["Day"] = df["DATE"].dt.day

df.drop(columns=["DATE", "IND", "IND.1", "IND.2"], inplace=True, errors="ignore")


X = df.drop(columns=["WIND"])
y = df["WIND"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(),
    "XGBoost": XGBRegressor(eval_metric='rmse')
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    results[name] = {
        "MAE": mean_absolute_error(y_test, predictions),
        "RMSE": mean_squared_error(y_test, predictions, squared=False),
        "R2 Score": r2_score(y_test, predictions)
    }


best_model_name = min(results, key=lambda x: results[x]["RMSE"])
best_model = models[best_model_name]


joblib.dump(best_model, "best_wind_model.pkl")
joblib.dump(scaler, "scaler.pkl")

print("\n🎯 Best Model Found:", best_model_name)
print("\n📁 Saved Files: best_wind_model.pkl & scaler.pkl")
print("\n📌 Model Results:")
print(pd.DataFrame(results))
```

```python
# ----------------------------
# Imports (all at top)
# ----------------------------
import pandas as pd
import joblib
from math import sqrt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Try import XGBoost, otherwise fallback
try:
    from xgboost import XGBRegressor
    xgb_available = True
except Exception:
    xgb_available = False
    # optional fallback (uncomment if you want)
    # from sklearn.ensemble import HistGradientBoostingRegressor

# ----------------------------
# 1) Load Dataset
# ----------------------------
df=pd.read_csv('wind_dataset.csv')

# ----------------------------
# 2) Fill missing values (mean for numeric)
# ----------------------------
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# ----------------------------
# 3) DATE -> features
# ----------------------------
df["DATE"] = pd.to_datetime(df["DATE"], errors='coerce')
df["Year"] = df["DATE"].dt.year
df["Month"] = df["DATE"].dt.month
df["Day"] = df["DATE"].dt.day

df.drop(columns=["DATE", "IND", "IND.1", "IND.2"], inplace=True, errors="ignore")

# ----------------------------
# 4) X, y
# ----------------------------
X = df.drop(columns=["WIND"])
y = df["WIND"]

# ----------------------------
# 5) Scale features
# ----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------------------
# 6) Train-test split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ----------------------------
# 7) Models dictionary
# ----------------------------
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42)
}

if xgb_available:
    models["XGBoost"] = XGBRegressor(eval_metric='rmse', use_label_encoder=False, random_state=42)
# else you may add a fallback model:
# models["HistGB"] = HistGradientBoostingRegressor(random_state=42)

# ----------------------------
# 8) Train & Evaluate
# ----------------------------
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)
    mse = mean_squared_error(y_test, preds)        # note: no 'squared' arg
    rmse = sqrt(mse)                               # RMSE computed manually
    r2 = r2_score(y_test, preds)

    results[name] = {"MAE": mae, "RMSE": rmse, "R2": r2}

# ----------------------------
# 9) Select best model by RMSE & Save model + scaler
# ----------------------------
best_model_name = min(results, key=lambda k: results[k]["RMSE"])
best_model = models[best_model_name]

joblib.dump(best_model, "best_wind_model.pkl")
joblib.dump(scaler, "scaler.pkl")

# ----------------------------
# 10) Print results
# ----------------------------
print("Best model:", best_model_name)
print(pd.DataFrame(results).T)
print("Saved: best_wind_model.pkl and scaler.pkl")
```
